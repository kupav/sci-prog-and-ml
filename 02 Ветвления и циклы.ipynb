# Задания по ветвлениям и циклам Python

## Задание 1: Чётное или нечётное
Напишите программу, которая запрашивает у пользователя целое число и выводит, является ли оно чётным или нечётным.

## Задание 2: Квадратное уравнение (простой случай)
Программа получает коэффициенты a, b, c для уравнения a*x² + b*x + c = 0. Вычислите дискриминант. Если он положительный, выведите сообщение "Два корня", если равен нулю — "Один корень", если отрицательный — "Действительных корней нет". Сами корни находить не нужно.

## Задание 3: Максимум из трёх
Запросите у пользователя три числа и выведите наибольшее из них, используя только ветвления (без функций max и сортировки).

## Задание 4: Високосный год
Год является високосным, если он кратен 4, но не кратен 100, либо кратен 400. Напишите программу, которая определяет по введённому году, високосный он или нет.

## Задание 5: Сумма чисел от 1 до N
Запросите у пользователя число N (натуральное). Вычислите и выведите сумму всех целых чисел от 1 до N.

## Задание 6: Факториал
Вычислите факториал введённого числа N (N! = 1*2*...*N). Учтите, что 0! = 1.

## Задание 7: Таблица умножения
Выведите на экран таблицу умножения (от 1 до 10) для числа, которое введёт пользователь.

## Задание 8: Сумма цифр
Запросите у пользователя целое положительное число. Найдите сумму его цифр, используя цикл while.

## Задание 9: Обратный отсчёт
Запросите число N. Затем с помощью цикла while выведите числа от N до 1, а в конце слово "Старт!".

## Задание 10: Поиск делителей
Выведите все натуральные делители введённого числа (кроме 1 и самого числа). Если таких делителей нет, выведите "Число простое".

## Задание 11: Проверка на простоту
Улучшите предыдущую задачу. Запросите число и определите, является ли оно простым. Выведите "Простое" или "Составное".

## Задание 12: Числа Фибоначчи
Выведите N первых чисел последовательности Фибоначчи. Напоминание: F₀=0, F₁=1, Fₙ = Fₙ₋₁ + Fₙ₋₂.

## Задание 13: Калькулятор с обработкой ошибок
Напишите простой калькулятор, который запрашивает два числа и операцию (+, -, *, /). Выполните операцию и выведите результат. Обработайте случай деления на ноль.

## Задание 14: Угадай число (компьютер загадывает)
Программа "загадывает" число от 1 до 100 (используйте фиксированное, например, 42). Пользователь пытается угадать. На каждый ход программа отвечает "Больше", "Меньше" или "Угадал!". Подсчитайте количество попыток.

## Задание 15: Числовая пирамида
Запросите число N (например, 5). Выведите пирамиду из чисел, используя вложенные циклы:
```
1
22
333
4444
55555
```

## Задание 16: Ряд Тейлора для e^x (приближённо)
Вычислите приближённое значение e^x по формуле: e^x ≈ 1 + x/1! + x²/2! + ... + xᴺ/N!. Запросите x и N (количество слагаемых).

## Задание 17: Переворот числа
Запросите целое положительное число. Выведите число, состоящее из тех же цифр, но записанных в обратном порядке, используя цикл while.

## Задание 18: Вывод всех чётных чисел в диапазоне
Запросите у пользователя два целых числа A и B (A < B). Выведите все чётные числа в этом диапазоне (включительно). Сделайте это двумя способами: с шагом в цикле и с условием внутри цикла.

## Задание 19: Сумма ряда с чередованием знаков
Вычислите сумму S = 1 - 1/2 + 1/3 - 1/4 + ... ± 1/N. Знак чередуется. Запросите N.

## Задание 20: График функции в консоли
Постройте упрощённый "график" функции y = x² в текстовом режиме для x от -M до M. Используйте вложенные циклы: по строкам (ось Y) и по столбцам (ось X). Отметьте точку '*', если y ≈ x² (с учётом масштаба и округления).
