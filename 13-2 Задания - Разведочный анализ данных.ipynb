# Практическое задание 2 по теме «Разведочный анализ данных (EDA)»

## Учебный датасет

Ниже создаётся искусственный датасет о пользователях онлайн-образовательной платформы.  
Целевая переменная: **completed_course** — завершил пользователь курс или нет.

В таблицу специально добавлены:
- пропуски;
- некорректные значения;
- разные записи одной и той же категории;
- выброс.

Это позволит потренировать базовые шаги EDA без сложных операций над данными.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display

np.random.seed(42)

n = 140

age = np.random.randint(18, 61, size=n).astype(float)
monthly_study_hours = np.clip(np.random.normal(24, 8, size=n), 2, 60)
monthly_fee = np.clip(np.random.normal(3200, 700, size=n), 500, 7000)
practice_tasks = np.random.randint(0, 31, size=n)
support_messages = np.random.poisson(2.2, size=n)
plan_type = np.random.choice(["Basic", "Pro", "Premium"], size=n, p=[0.45, 0.35, 0.20]).astype(object)
device_type = np.random.choice(["Web", "Mobile", "Tablet"], size=n, p=[0.45, 0.45, 0.10]).astype(object)

score = (
    0.05 * monthly_study_hours
    + 0.09 * practice_tasks
    - 0.22 * support_messages
    + 0.00025 * monthly_fee
    + (plan_type == "Premium") * 0.45
    + (plan_type == "Pro") * 0.20
    + np.random.normal(0, 0.45, size=n)
)
completed_course = (score > 2.6).astype(int)

df_source = pd.DataFrame({
    "age": age.round(0),
    "monthly_study_hours": monthly_study_hours.round(1),
    "monthly_fee": monthly_fee.round(0),
    "practice_tasks": practice_tasks,
    "support_messages": support_messages,
    "plan_type": plan_type,
    "device_type": device_type,
    "completed_course": completed_course
})

# Добавляем искажения в данные
df_source.loc[[5, 48], "age"] = -3
df_source.loc[[12, 87], "monthly_fee"] = np.nan
df_source.loc[33, "monthly_fee"] = -1500
df_source.loc[71, "monthly_study_hours"] = 130
df_source.loc[[19, 52, 121], "device_type"] = np.nan
df_source.loc[[8, 63], "plan_type"] = "pro"
df_source.loc[[26, 118], "plan_type"] = "Premium "
df_source.loc[99, "support_messages"] = 45

df_source.to_csv("eda_practice_dataset_2.csv", index=False)

print("Файл eda_practice_dataset_2.csv создан.")
display(df_source.head())
````

# Задание 10. Сравнение модели на «сыром» и очищенном датасете

## Что нужно сделать
1. Попробуйте обучить простую модель на исходном датасете `df`.
2. Затем обучите такую же модель на очищенном датасете `df_clean`.
3. Сравните результаты и сделайте вывод, зачем нужен EDA.

## Зачем это нужно
Это итоговое задание, где видно, что EDA влияет не только на аккуратность таблицы, но и на возможность построить модель.

Логика такая:
- сначала пробуем обучить модель на данных «как есть»;
- затем используем очищенные данные;
- сравниваем, изменилось ли обучение и качество.

## Почему это важно
Такой эксперимент показывает, что EDA:
- делает обучение модели возможным;
- уменьшает число технических ошибок;
- помогает получить более понятные признаки.

## Подсказки по Pandas и машинному обучению
- признаки `X` нужно отделить от целевой переменной `y`;
- категориальные признаки можно перевести в числа через `pd.get_dummies(...)`;
- данные обычно делят на обучающую и тестовую части;
- после этого создают модель, обучают её и оценивают качество.

Если вы ещё не изучали модели, используйте такой шаблон:

1. Подготовка признаков и целевой переменной:
```python
X = df_clean.drop(columns=["completed_course"])
y = df_clean["completed_course"]
```

2. Преобразование категориальных столбцов:
```python
X = pd.get_dummies(X, drop_first=False)
```

3. Разбиение на обучающую и тестовую части:
```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

4. Создание и обучение модели:
```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```
Увеличьте `max_iter` если модель не будет успевать сойтись.

5. Получение предсказаний и оценка качества:
