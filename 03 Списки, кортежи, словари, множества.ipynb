# Задания по структурам данных Python

## Задание 1: Создание и базовые операции со списками
Создайте список `prime_numbers` с первыми пятью простыми числами. Добавьте в конец списка число 11. Вставьте число 13 на позицию с индексом 5. Удалите число 3 из списка. Выведите итоговый список и его длину.

## Задание 2: Срезы (slicing) списка
Дан список `numbers = list(range(1, 21))` (числа от 1 до 20). Создайте:
1. Список `first_half` — первые 10 элементов
2. Список `even_numbers` — все четные числа
3. Список `reversed_list` — список в обратном порядке
4. Список `every_third` — каждый третий элемент, начиная со второго

Выведите все созданные списки.

## Задание 3: Генератор списков (list comprehension) для математической последовательности
Создайте список квадратов первых 10 натуральных чисел, используя генератор списка. Затем создайте список кубов только четных чисел от 1 до 10. Выведите оба списка.

## Задание 4: Работа с кортежами и распаковка
Создайте кортеж `vector` с тремя координатами (x, y, z). Распакуйте кортеж в три отдельные переменные. Затем создайте новый кортеж `vector2`, который будет результатом покоординатного сложения `vector` с кортежем (1, 2, 3). Выведите оба кортежа и распакованные координаты.

## Задание 5: Множества и операции над ними
Создайте два множества: `A = {1, 2, 3, 4, 5, 6}` и `B = {4, 5, 6, 7, 8, 9}`. Вычислите и выведите:
1. Объединение A и B
2. Пересечение A и B
3. Разность A - B
4. Симметрическую разность (элементы, принадлежащие только одному из множеств)

Проверьте, является ли множество `{2, 4}` подмножеством A.

## Задание 6: Удаление дубликатов через множество
Дан список с повторяющимися элементами: `data = [1, 2, 2, 3, 4, 4, 4, 5, 3, 2, 1]`. Преобразуйте его в множество, чтобы удалить дубликаты. Затем преобразуйте обратно в список и отсортируйте по возрастанию. Выведите исходный список, множество и итоговый отсортированный список.

## Задание 7: Словарь для хранения информации о функциях
Создайте словарь `functions`, где ключами будут названия математических функций (`'квадрат'`, `'куб'`, `'корень'`), а значениями — лямбда-функции, вычисляющие соответствующие значения. Добавьте к словарю функцию `'синус'`, используя `math.sin`. Затем вычислите значения всех функций для аргумента `x = 2.5` и выведите результат.

## Задание 8: Сортировка списка кортежей
Дан список кортежей вида `(имя, балл1, балл2)`. Отсортируйте этот список:
1. По убыванию суммы баллов (балл1 + балл2)
2. При равных суммах — по имени в алфавитном порядке

Выведите отсортированный список.

## Задание 9: Глубокое копирование вложенных структур
Создайте вложенный список (матрицу 2x2). Сделайте поверхностную копию через `copy()` и глубокую копию через `deepcopy()`. Измените элемент в исходной матрице. Выведите все три матрицы, чтобы увидеть разницу.

## Задание 10: Поиск в списке и словаре
Создайте список из 100 случайных целых чисел от 1 до 10. Создайте словарь, где ключи — числа от 1 до 10, а значения — количество раз, которое это число встречается в списке. Найдите число, которое встречается чаще всего, и выведите его и частоту.

## Задание 11: Множества для работы с уникальными элементами строк
Дана строка: `"математика и информатика"`. Создайте множество уникальных букв в этой строке (игнорируйте пробелы). Затем создайте второе множество для строки `"алгоритмы"`. Найдите общие буквы в обеих строках (пересечение множеств). Выведите результаты.

## Задание 12: Работа с zip и enumerate
Даны два списка: `coeffs = [2, -3, 1]` (коэффициенты полинома `2x² - 3x + 1`) и `powers = [2, 1, 0]` (степени x). С помощью `zip` создайте список строк вида `"2*x^2"`, `"-3*x^1"`, `"1*x^0"`. Затем с помощью `enumerate` выведите полином в удобочитаемом виде, пропуская слагаемые с нулевым коэффициентом.

## Задание 13: Словарь с ключами-кортежами (работа с точками)
Создайте словарь `points`, где ключами будут кортежи `(x, y)` — координаты точек, а значениями — расстояние от этой точки до начала координат. Добавьте точки (0,0), (3,4), (1,1), (5,12). Выведите словарь, отсортированный по возрастанию расстояния.

## Задание 14: Множества и логические условия
Создайте множество `perfect_squares` квадратов чисел от 1 до 20. Создайте множество `even_numbers` четных чисел от 1 до 30. Найдите:
1. Числа, которые являются одновременно четными и квадратами (пересечение)
2. Четные числа, которые не являются квадратами (разность)
3. Числа, которые являются либо квадратами, либо четными, но не одновременно (симметрическая разность)

Выведите размер каждого полученного множества.

## Задание 15: Вложенные словари (словарь словарей)
Создайте словарь `students`, где ключи — ID студентов, а значения — вложенные словари с информацией: имя, фамилия, оценки (список из 3 оценок). Добавьте 3 студентов. Затем для каждого студента вычислите средний балл и добавьте его во вложенный словарь с ключом `'средний'`. Выведите итоговый словарь.

## Задание 16: Фильтрация списка с условием
Дан список чисел от 1 до 50. Создайте новый список, содержащий только числа, которые:
1. Делятся на 3 ИЛИ на 5
2. При этом НЕ делятся на 15

Используйте генератор списка. Выведите первые 10 элементов полученного списка.

## Задание 17: Объединение и преобразование коллекций
Даны два списка: `keys = ['a', 'b', 'c', 'd']` и `values = [1, 2, 3, 4]`. Создайте словарь с помощью `zip`. Затем создайте список кортежей `(key, value)` из этого словаря. Из этого списка создайте множество, если сумма key и value четная (используйте генератор множества). Выведите все промежуточные результаты.

## Задание 18: Словарь с вычисляемыми значениями
Создайте словарь `math_constants` с ключами-строками и значениями-числами (например, π, e, φ). Затем создайте новый словарь `inverse_constants`, где ключи те же, а значения — обратные величины (1/x). Учтите возможность деления на ноль. Выведите оба словаря.

## Задание 19: Работа с неизменяемыми коллекциями
Создайте кортеж из трех элементов, где:
1. Первый элемент — число
2. Второй элемент — кортеж из двух чисел
3. Третий элемент — список (изменяемый!)

Попробуйте изменить список внутри кортежа. Затем попробуйте изменить сам кортеж (переприсвоить элемент). Объясните, что работает, а что нет, и почему.

## Задание 20: Сложная структура для статистики
Создайте структуру данных, которая хранит оценки студентов по разным предметам. Используйте словарь, где ключ — имя студента, значение — другой словарь (предмет → список оценок). Заполните данными для 2-3 студентов по 2-3 предметам. Напишите код, который для каждого студента вычисляет средний балл по каждому предмету и общий средний балл.
