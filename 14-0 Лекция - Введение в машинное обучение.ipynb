## Классическая наука и машинное обучение

Вся наука связана с предсказанием будущего.

Прежде всего собирают данные из интересующей области: например, фиксируют положения планет в зависимости от времени года, температуру воздуха в зависимости от скорости ветра, дальность полёта пушечного ядра в зависимости от количества пороха и так далее.

Затем пытаются выявить зависимости: действительно ли одна величина зависит от другой? Какой математической формулой можно описать эту зависимость?

Следующий шаг — проверить найденные математические формулы на новых наблюдениях. Действительно ли они работают? После этого появляются экспериментальные законы: второй закон Ньютона и закон Кулона в физике, менделевские закономерности наследования в генетике и так далее.

Когда открыто достаточно большое число законов, возникает математическая теория, которая объединяет их и даёт инструменты для предсказаний.

Примеры предсказаний, выполняемых таким образом: морская навигация (по наблюдению звёзд предсказывают положение корабля), безопасная нагрузка на мост (её можно вычислить, зная свойства материалов), вероятность генетических заболеваний (с использованием законов генетики).

Главная проблема построения науки таким способом состоит в том, чтобы найти наиболее простые, фундаментальные законы управления, лежащие в основе всех остальных законов.

В физике таким фундаментальным законом, например, являются законы движения Ньютона, в генетике — знание структуры ДНК и так далее.

И закон есть закон: явления всегда происходят так, как он утверждает. Без исключений.

Успех всей области зависит от того, насколько успешно удаётся найти такие базовые законы.

Всё это можно назвать классическим подходом к созданию знания.

Во второй половине XX века, когда компьютеры позволили работать с большими объёмами данных, учёные начали пытаться применять классические идеи к более сложным областям, например к социальным отношениям или медицине.

Но в таких областях фундаментальные законы обнаружить не удаётся. Удаётся выявить лишь некоторые вероятностные зависимости.

Если у вас есть большой практический опыт в некоторой области, скажем в медицине, то вы являетесь экспертом.

Возможно, вы не знаете точных фундаментальных законов (потому что их никто не знает), но вы уверены, что за событием A с высокой вероятностью последует событие B.

Например, врач, наблюдая определённые симптомы, может быть уверен, что у пациента есть определённое заболевание.

Как формализовать этот опыт, как извлечь знания у эксперта и поместить их в компьютер?

Для этого были созданы компьютерные экспертные системы. В основном это были списки вручную закодированных правил вида "if then else".

Эти правила создавались после опроса экспертов в соответствующей области.

По сути это была лишь попытка применить классические методы там, где они плохо работают: вспомним, что классический научный подход требует перехода от сырых наблюдений к извлечению законов. Но компьютерные экспертные системы застряли на первом шаге.

Две основные проблемы компьютерных систем, полностью основанных на экспертных правилах, таковы:

- Отсутствие обобщения. Мы можем отвечать только на те вопросы, которые были заданы эксперту. Никакие законы не выводятся, поэтому ранее не встречавшиеся случаи не могут быть корректно обработаны.

- Отсутствие универсальности. Такие системы решают только узкие задачи и не способны решать другие. Компьютерные системы, основанные на мнении экспертов, не могли распознавать лица на фотографиях. Причина в том, что ни один человек-эксперт не может чётко сформулировать строгие правила, позволяющие ему это делать.

Именно поэтому появилось понятие машинного обучения.

**Машинное обучение — это совокупность математических методов и компьютерных алгоритмов, которые автоматически извлекают знания из данных.**

Ключевые требования здесь — обобщение и универсальность.

Извлечённое знание должно быть общим.

Это означает, что метод машинного обучения не может просто запоминать ответы.

Он должен выводить общие зависимости в данных и уметь корректно обрабатывать новые, ранее не встречавшиеся случаи.

Именно это делает классическая наука, извлекая фундаментальные законы из экспериментальных данных.

Есть два различия.

- Законы классической науки открываются людьми. Обобщение данных выполняется автоматически компьютерной программой в процессе обучения.

- Поскольку классические законы открываются людьми, они имеют форму, понятную человеку. Эти законы представлены в виде математических уравнений или текста на естественном языке. Знания машинного обучения, полученные после обобщения данных, состоят из чисел. Обычно это огромные массивы чисел. В большинстве случаев человек не может понять это знание. Мы можем лишь использовать его, не понимая, почему оно работает.

Методы машинного обучения универсальны. Они применимы в любой области, которую можно описать набором данных.

Конечно, одни методы лучше работают для распознавания лиц, а другие предпочтительнее для прогнозирования временных рядов.

Но ключевые идеи являются общими, и любая интересующая область может быть проанализирована методами машинного обучения.


## Машинное обучение

Машинное обучение связано с извлечением знаний из данных.

В последние годы применение методов машинного обучения стало повсеместным в повседневной жизни.

От автоматических рекомендаций, какие фильмы посмотреть, какую еду заказать или какие товары купить, до персонализированного интернет-радио и распознавания друзей на ваших фотографиях — во многих современных сайтах и устройствах алгоритмы машинного обучения лежат в основе работы.

За пределами коммерческих приложений машинное обучение оказало огромное влияние на то, как сегодня ведутся исследования, основанные на данных.

Инструменты машинного обучения были применены к самым разным научным задачам, таким как изучение звёзд, поиск далёких планет, открытие новых частиц, анализ последовательностей ДНК и подбор персонализированного лечения рака.

Возможно, самая важная часть процесса машинного обучения — это понимание данных, с которыми вы работаете, и того, как они связаны с задачей, которую вы хотите решить.

Прежде чем начинать строить модель, необходимо понять, что именно происходит в наборе данных и сообразно этому выбрать модель. 

Каждая машинная модель отличается тем, для какого типа данных и для какой постановки задачи он подходит лучше всего.

Когда вы строите решение на основе машинного обучения, следует ответить — или по крайней мере держать в уме — следующие вопросы:

- На какой(-ие) вопрос(-ы) я пытаюсь ответить? Думаю ли я, что собранные данные позволяют ответить на этот вопрос?
- Как лучше всего сформулировать мой(-и) вопрос(-ы) как задачу машинного обучения?
- Достаточно ли я собрал данных, чтобы представить проблему, которую хочу решить?
- Какие признаки данных я извлёк, и позволят ли они делать правильные предсказания?
- Как я буду измерять успех в моём приложении?
- Как решение на основе машинного обучения будет взаимодействовать с другими частями моего исследования или бизнес-продукта?

В более широком контексте алгоритмы и методы машинного обучения — это лишь часть общего процесса решения конкретной задачи, и полезно всегда помнить о полной картине.

Многие люди тратят много времени на построение сложных решений в области машинного обучения, а потом обнаруживают, что решали не ту задачу.

Углубляясь в технические аспекты машинного обучения легко потерять из виду конечные цели.

## Моделирование

Ключевую роль в машинном обучении играют модели.

Модели являются результатом извлечения знаний и способны предсказывать будущее.

Что такое модель?

Это по сути описание математической зависимости, существующей между различными переменными.

Например, если вы пытаетесь привлечь деньги для своего сайта социальной сети, вы можете построить бизнес-модель (вероятно, в электронной таблице), которая принимает на вход такие величины, как "число пользователей", "доход от рекламы на пользователя" и "число сотрудников", и выдаёт вашу годовую прибыль на ближайшие несколько лет.

Рецепт из поваренной книги тоже задаёт модель, связывающую такие входы, как "число едоков" и "степень голода", с количеством необходимых ингредиентов.

А если вы когда-нибудь смотрели покер по телевизору, то знаете, что "вероятность выигрыша" каждого игрока оценивается в реальном времени на основе модели, которая учитывает уже открытые карты и распределение карт в колоде.

Бизнес-модель, вероятно, основана на простых математических соотношениях: прибыль — это выручка минус расходы, выручка — это число проданных единиц, умноженное на среднюю цену, и так далее.

Модель рецепта, вероятно, основана на методе проб и ошибок: кто-то пришёл на кухню и пробовал разные комбинации ингредиентов, пока не нашёл понравившийся вариант.

А покерная модель основана на теории вероятностей, правилах покера и некоторых достаточно безобидных предположениях о случайном процессе раздачи карт.

Более строго, модель — это упрощённая система, обладающая наиболее важными свойствами и отношениями другой системы.

Модели создаются с использованием формального языка, такого как математическая запись или язык программирования.

Естественный язык в целом плохо подходит для построения моделей из-за своей расплывчатости. Однако иногда он всё же используется, например в психологии.

Модели создаются для предсказания будущего моделируемой системы.

Обычно природные системы (например, социальные) слишком сложны для прямого изучения.

В этом случае создаётся модель, сохраняющая только наиболее существенные черты исходной системы.

Изучение модели даёт знания о поведении исходной системы в будущем.

Поскольку модели всегда урезаны по числу признаков по сравнению с исходной системой, ошибки неизбежны.

Построение модели всегда является компромиссом между простотой (а значит, возможностью её изучения) и точностью.

## Модели машинного обучения

Модель машинного обучения — это совокупность компьютерной программы и массива параметров (то есть обычных компьютерных переменных).

У модели есть входы, куда подаётся информация из интересующей области, и выходы, на которых появляется предсказание.

Предсказания формируются после обработки входов компьютерной программой модели с использованием параметров.

Хотя сама компьютерная программа часто бывает стандартной, значения параметров специально настраиваются в процессе обучения.

Параметры — самая ценная часть модели. Их число обычно велико, и они хранятся во внешней памяти в файлах.

Хотя предсказания выполняет сама модель, для её обучения требуется дополнительное программное обеспечение.

Оно может быть стандартным или написанным специалистом по анализу данных под конкретную задачу.

Обучающее программное обеспечение

- выполняет сбор данных из различных источников (data mining, big data),
- обеспечивает доступ к данным (чтение из файлов, загрузку из веба и т. д.),
- подаёт данные в модель,
- получает предсказание,
- оценивает его качество,
- корректирует параметры модели, чтобы при необходимости улучшить её.

## Обучение с учителем

Эти модели обучаются на размеченном наборе данных.

Это означает, что каждому входу, то есть каждой записи данных, сопоставлен желаемый выход.

В процессе обучения вход подаётся в модель.

Модель формирует выход, который сравнивается с желаемым.

Если они различаются, параметры модели корректируются так, чтобы уменьшить различие.

Размеченные наборы данных создаются людьми. Часто это делается добровольцами через краудсорсинговые платформы.

Известный пример — ImageNet.

Проект ImageNet представляет собой большую визуальную базу данных, предназначенную для исследований в области программ распознавания объектов.

Более 14 миллионов изображений были вручную размечены в рамках проекта, чтобы указать, какие объекты на них изображены. По крайней мере для одного миллиона изображений также указаны ограничивающие рамки.

Цель обучения с учителем — извлечение и сохранение существенных признаков данных в виде параметров модели.

Желаемый результат обучения: алгоритм способен сформировать правильный выход для входа, которого он никогда раньше не видел, без какой-либо помощи человека.

Например, можно создать фильтр спама, который анализирует содержимое электронных писем и решает, является ли письмо спамом.

Чтобы обучить его, нужно создать набор данных, включающий различные письма, помеченные как спам или не спам.

После обучения, получив новое письмо, алгоритм сможет предсказать, является ли оно спамом.

Другие примеры задач машинного обучения с учителем:

- Идентификация пользователя по лицу.
- Определение почтового индекса по рукописным цифрам на конверте.
- Выявление мошеннической активности в транзакциях по кредитным картам.
- Распознавание номера автомобиля на изображениях с дорожных камер.


## Обучение без учителя

В обучении без учителя известны только входные данные, а никакие известные выходные данные модели не предоставляются.

Цель обучения без учителя — обнаружение закономерностей в данных.

Один из типов задач обучения без учителя состоит в том, чтобы обнаруживать группы похожих примеров внутри данных. Это называется __кластеризацией данных__.

Также этот класс задач называют __непараметрическим обучением без учителя__ (в сравнении со следующим типом).

Второй тип известен как __оценивание плотности__ или __параметрическое обучение__.

В этом случае предполагается, что данные не разбиваются на кластеры, а распределены в соответствии с некоторым законом, например нормальным распределением.

Мы предполагаем некоторую функцию распределения, которой подчиняются данные, и цель состоит в вычислении её параметров.

Примеры задач обучения без учителя:

- Выявление тем в наборе сообщений блога
- Сегментация клиентов на группы со схожими предпочтениями
- Обнаружение аномальных шаблонов доступа к веб-сайту


## Обучение с подкреплением

Эти модели в некоторой степени похожи на модели обучения с учителем. Но между ними всё же есть серьёзное различие.

Если обучение является обучением с учителем, то мы знаем глобальную функцию потерь (функцию ошибки, т.е. функцию, которая она показывает, наскольо каждый отрвет модели отличается от точного).

Это означает, что мы всегда знаем точный желаемый выход для каждого обучающего входа.

Следовательно, мы можем сравнить желаемый и полученный выходы и вычислить расстояние между ними.

Зная это расстояние, мы используем алгоритм оптимизации (например, градиентный спуск) и напрямую вычисляем обновления параметров в направлении оптимума.

В процессе обучения с подкреплением модель взаимодействует с динамической средой, которая подаёт ей входы, а модель отвечает выходами (примеры — управление транспортным средством или игра против соперника).

Отличие от обучения с учителем заключается в том, что точный желаемый выход неизвестен.

Среда является динамической, то есть ситуация меняется во времени.

Тренер модели не может глобально оценить её выход (приводит ли этот ход к победе или к поражению), но может сказать, хорош он или плох в данной конкретной ситуации.

Поэтому каждый выход модели оценивается в терминах наград и штрафов, а обновления параметров вычисляются так, чтобы уменьшать штрафы и увеличивать награды в будущих ситуациях.

Обучение с подкреплением используется для того, чтобы обучить компьютер играть в шахматы или в игру го.

Фактически компьютер обучается сам.

Процесс обучения начинается, и модель делает ходы за одну и за другую сторону.

Обучающее программное обеспечение оценивает её ходы с использованием известных функций, характеризующих качество игровой позиции.

На основе этих оценок параметры модели обновляются.

После достаточно большого числа партий модель становится гроссмейстером.


## Переобучение и недообучение

Стандартная проблема в машинном обучении — переобучение.

Переобученная модель очень хорошо работает на обучающих данных, но терпит неудачу на любых новых данных.

Переобучение означает, что вместо нахождения обобщения обучающих данных модель просто запомнила их.

Противоположная проблема — недообучение.

Недообученная модель плохо работает даже на обучающих данных.

Переобучение обычно возникает, когда модель слишком сложна для данного набора данных, то есть имеет слишком много настраиваемых параметров.

Недообучение, напротив, означает, что модель слишком проста. Числа её степеней свободы (параметров) недостаточно, чтобы хранить извлечённую информацию.

Другая причина состоит в том, что в наборе данных недостаточно признаков (столбцов), и он плохо описывает решаемую задачу.

Борьба с недообучением и переобучением — центральная проблема создания модели машинного обучения.

Когда модель недообучена, нужно увеличить её ёмкость.

Это означает, что её структуру нужно изменить так, чтобы добавить больше параметров.

Можно добавить в модель больше структурных элементов (например, больше слоёв в нейронную сеть), или можно собрать больше признаков из интересующей области.

Простой приём — добавить новые столбцы данных как степени старых: квадраты, кубы и так далее.

Если же возникает переобучение, то можно либо упростить модель, уменьшив число её параметров, либо попытаться найти больше записей данных (хотя это часто невозможно), чтобы увеличить обучающий набор данных.

Ещё один способ борьбы с переобучением называется регуляризацией.

Регуляризация означает ограничение вариации параметров.

У переобученных моделей параметры часто имеют очень большие по модулю значения.

Когда мы применяем регуляризацию, мы добавляем штраф за величину параметров: чем они меньше, тем лучше.

И ещё один способ победить переобучение называется dropout.

В процессе обучения мы случайным образом отключаем некоторые параметры модели, то есть исключаем их из процедуры обучения.

Как мы видим, все способы борьбы с переобучением направлены на то, чтобы не дать модели просто запомнить набор данных.


## Разбиение набора данных на обучающую, валидационную и тестовую части

Недообучение видно сразу: модели просто работают плохо.

А переобучение нельзя обнаружить, рассматривая только обучающий набор данных.

Чтобы выявить переобучение, нужно разделить набор данных на обучающую и тестовую части.

Но здесь возникает другая проблема.

Представим, что мы выполнили несколько шагов обучения и хотим посмотреть, как идёт процесс.

Мы берём тестовую часть данных и оцениваем качество модели.

Если оно такое же хорошее, как и на обучающих данных, всё хорошо, переобучения нет.

Но если модель работает плохо на тестовых данных, мы обнаруживаем переобучение.

В этом случае мы что-то меняем в модели.

Это означает, что тестовые данные тоже участвуют в обучении.

Они влияют, пусть и косвенно, на модификацию модели.

Но итоговая оценка модели может быть получена только на данных, которые модель никогда не видела и которые никак не влияли на её обучение.

Это означает, что разбиения исходного набора данных только на обучающую и тестовую части недостаточно.

Мы должны разделить набор данных на три части: обучающую, валидационную и тестовую:
- обучающие данные непосредственно используются для обновления параметров модели;
- валидационные данные используются после нескольких шагов обучения для оценки качества и переобучения модели; на основе этой оценки в процесс обучения могут вноситься корректировки;
- тестовые данные используются только один раз для получения итоговой оценки модели; после этого никакие обновления параметров модели выполняться не должны.

Обычно набор данных делят в пропорции 70\%, 20\%, 10\% или 80\%, 10\%, 10\%.

Обучающая часть обычно больше всех остальных.


## Кросс-валидация

Если размер всего набора данных не слишком велик, мы можем улучшить обучение, используя кросс-валидацию.

Сначала мы откладываем тестовые данные, чтобы получить итоговую оценку модели.

Остальные данные не делятся на обучающие и валидационные части навсегда.

Вместо этого мы разбиваем все оставшиеся данные на $k$ частей. Сначала $k-1$ частей используются для обучения модели, а затем её качество проверяется на последней части.

Затем мы переставляем части данных так, чтобы следующая часть стала тестовой.

Процедура повторяется $k$ раз, а результаты тестов усредняются.

Такая скользящая оценка позволяет более равномерно использовать данные.

Подчеркнём: кросс-валидация - это не лучшая практика, а компромисс в условиях, когда набор данных небольшой и жалко выделять для валидации его значительный фрагмент. Лучшая практика - выделить валидационные данные на постоянной основе и никогда не использовать их для на этапе подстройки параметров. Они нужны только чтобы проверить насколько хорошо работает модель на текущем этапе обучения.

## Стохастический градиентный спуск, эпоха

Часто набор данных очень велик, и у компьютера не хватает памяти, чтобы хранить его целиком.

В этом случае набор данных делят на части, называемые мини-пакетами (minibatches) или просто пакетами (batches).

Вместо применения алгоритма обучения ко всему набору данных сразу его применяют к мини-пакетам, а параметры модели обновляются после каждого мини-пакета.

Обычно это используется тогда, когда параметры оптимизируются методом градиентного спуска.

Градиентный спуск, применяемый к мини-пакетам, называется стохастическим градиентным спуском.

Его можно рассматривать как стохастическое приближение градиентного спуска, поскольку он заменяет истинный градиент (вычисленный по всему набору данных) его оценкой (вычисленной по случайно выбранному подмножеству данных).

Особенно в задачах оптимизации большой размерности это уменьшает вычислительную нагрузку, обеспечивая более быстрые итерации в обмен на меньшую скорость сходимости.

Когда набор данных разбит на мини-пакеты и мы по одному подали их все в процедуру обучения, это называется эпохой.

Иными словами, эпоха проходит тогда, когда мы показали модели все мини-пакеты, то есть весь обучающий набор данных.


## Регрессия и классификация

Обучение с учителем обычно решает две основные задачи:

- регрессию,
- классификацию.

Классификация — это задача присвоения меток объектам данных, принадлежащим к различным классам. Например, обучение модели различать кошек и собак — это задача классификации, где кошки и собаки являются двумя классами.

Регрессия, в свою очередь, — это задача предсказания непрерывных значений на основе различных независимых признаков. Например, прогнозирование цены дома по таким признакам, как число спален, район и т. д.


## Метрики для регрессии

Качество моделей машинного обучения оценивается с помощью метрик.

В задачах регрессии у нас есть истинный вектор $\hat y$ и вектор $y$, предсказанный моделью.

Метрики показывают, насколько они различаются.

Обычно используется та или иная версия расстояния.

Предположим, что векторы $\hat y$ и $y$ имеют $n$ элементов.

Например, у следующих векторов $n=5$:
$$
\hat y = (2.1, 3.4, -5.1, 1.6, -3.3)
$$
$$
y = (2.2, 3.45, -5.0, 1.7, -3.2)
$$

Обозначим элементы векторов как $\hat y_i$ и $y_i$, где $i=1,2,\ldots, n$.

Для примера выше $y_1=2.2$, $y_2=3.45$.

__MSE, среднеквадратичная ошибка (Mean Squared Error)__, наиболее часто используемая метрика:
$$
	\text{MSE} = \frac{1}{n}\sum_{i=1}^n (\hat y_i - y_i)^2
$$
Это квадрат евклидова расстояния между двумя векторами, делённый на $n$.

__RMSE, корень из среднеквадратичной ошибки (Root Mean Squared Error)__. Это квадратный корень из MSE, то есть евклидово расстояние, делённое на $\sqrt{n}$.
$$
	\text{RMSE} = \sqrt{\frac{1}{n}\sum_{i=1}^n (\hat y_i - y_i)^2}
$$

__MAE, средняя абсолютная ошибка (Mean Absolute Error)__.
$$
	\text{MAE} = \frac{1}{n}\sum_{i=1}^n |\hat y_i - y_i|
$$
Это манхэттенское расстояние, делённое на $n$.

Различие между этими метриками состоит в разном масштабе для малых и больших ошибок.

Например, MSE возводит ошибку в квадрат, поэтому сильнее штрафует большие ошибки по сравнению с RMSE.

А когда ошибки малы, RMSE штрафует их сильнее, чем MSE.

MAE, в отличие от этих двух метрик, даёт одинаковые штрафы и для малых, и для больших ошибок.


## Метрики для классификации

Когда модель выполняет классификацию, она должна вернуть метку класса.

Но обычно фактическим выходом модели является вектор $y$ из $n$ элементов, где $n$ — число классов.

Чтобы получить метку класса, этот вектор нормируют. Часто используется так называемая функция softmax:
$$
q_i = \frac{e^{y_i}}{\sum_{i=1}^n e^{y_i}}
$$

После нормировки все элементы вектора становятся положительными, а их сумма равна 1. Если один из элементов стремится к 1, все остальные стремятся к нулю.

Предсказанная метка класса соответствует тому элементу вектора, который после нормировки имеет наибольшее значение.

Чтобы оценить, насколько хорошим является предсказание, этот вектор сравнивают с желаемым предсказанием.

Для этого желаемые метки классов представляют в one-hot форме:
$$
p=(1,0,0,0,0) 	\text{ : class 1}
$$
$$
p=(0,1,0,0,0) 	\text{ : class 2}
$$
$$
p=(0,0,1,0,0) 	\text{ : class 3}
$$
и так далее.

Затем предсказание модели $q$ сравнивают с желаемым one-hot-выходом с помощью __кросс-энтропии__:
$$
H(p,q) = - \sum_{i=1}^n p_i \log q_i
$$

Эта кросс-энтропия будет очень большой, если модель предсказывает неверный класс (то есть наибольшим оказывается неправильный элемент вектора), и становится очень малой, когда правильный элемент предсказанного вектора намного больше остальных.

В случае бинарной классификации модель возвращает только одно вещественное значение $y$.
Прежде всего его нужно отобразить в диапазон $[0,1]$.

Обычно используется сигмоидная функция $\sigma(y)$:
$$
q = \sigma(y) = \frac{1}{1+e^{-y}}
$$
Выход этой функции принадлежит диапазону $[0,1]$.


```python
import numpy as np
import matplotlib.pyplot as plt
xs = np.linspace(-5, 5, 100)
ys = 1 / (1 + np.exp(-xs))  # The sigmoid function is computed here
fig, ax = plt.subplots()
ax.plot(xs, ys)
ax.grid()
ax.set_title("Sigmod function")
ax.set_xlabel("y")
ax.set_ylabel("s");
```

Метки, то есть желаемые выходы $p$ в бинарной классификации,
принимают значения либо 0, либо 1.

Таким образом, зная желаемую метку $p$ и предсказанное значение $q$,
вычисляют __бинарную кросс-энтропию__:
$$
H(p,q) = -p \log q - (1-p) \log (1-q)
$$


## Метрики для бинарной классификации: accuracy, precision и recall. Коэффициент корреляции Мэттьюса

При работе с задачей классификации на первый взгляд кажется очевидным оценивать качество по accuracy, то есть по доле правильных предсказаний.

Но на самом деле это плохая идея, когда данные несбалансированы, то есть когда категории встречаются с разными вероятностями.

Для простоты рассмотрим бинарную классификацию:
является ли это письмо спамом или нет, болен ли этот человек или нет, будет ли завтра дождь и так далее.

Существует два типа предсказаний: положительное и отрицательное. И каждое из них может быть истинным или ложным.
Мы знаем это, поскольку рассматриваем обучение с учителем и имеем истинные метки.

- Истинно положительный результат (true positive): модель говорит, что это сообщение — спам, и это действительно спам.
- Ложно положительный результат (false positive): модель говорит, что это спам, но на самом деле это не спам. Ошибка I рода, ложное обнаружение.
- Ложно отрицательный результат (false negative): модель говорит, что сообщение не является спамом, но это неверно. Ошибка II рода, пропуск обнаружения.
- Истинно отрицательный результат (true negative): модель говорит, что сообщение не является спамом, и это действительно не спам.

Все эти случаи можно собрать в таблицу, которая называется матрицей ошибок.

|                    | На самом деле спам                      | На самом деле не спам                  |
|--------------------|-----------------------------------------|----------------------------------------|
| Предсказан спам    | TP (число <br>истинно положительных случаев)  | FP (число<br>ложноположительных случаев) |
| Предсказан не спам | FN (число <br>ложноотрицательных случаев) | TN (число <br>истинно отрицательных случаев) |

Теперь создадим спам-фильтр, который будет иметь accuracy 90%. Это означает, что 90% его предсказаний будут правильными.

Фильтр будет выдавать предсказания случайным образом.

В большинстве случаев он будет помечать сообщения как не спам и
только с вероятностью $p=0.001$ будет сообщать о спаме.

Смоделируем его работу. Сначала создадим сообщения.

Предположим, что одно спам-сообщение приходится на каждые 10 сообщений. Наш набор данных будет содержать $10^6$ сообщений.


```python
import numpy as np
rng = np.random.default_rng()

# One spam message (marked as 1) for each 10 messages (marked as 0), 1000000 totqly
messages = ([1] + [0] * 9) * 100000

# Shuffle spam and non-spam messages
rng.shuffle(messages)
```

Теперь создадим фильтр. Он просто генерирует 1 (это означает, что предсказан спам) с малой вероятностью.


```python
p_spam_filter = 0.001

# Do predicitons without even looking at messages
predicts = [1 if rng.random() < p_spam_filter else 0 for _ in range(len(messages))]
```

Вычислим accuracy нашего фильтра: пройдём по сообщениям и предсказаниям параллельно и поставим 1, если они совпадают.


```python
# Put 1 when precition is correct
success = [1 if m == p else 0 for m, p in zip(messages, predicts)]

accuracy = sum(success) / len(messages)
print(f"accuracy={accuracy:.3f}")
```

Наш случайный фильтр предсказывает правильные результаты в 90% случаев.

Почему?

Список сообщений содержит много нулей (сообщений не спам). И спам-фильтр в основном тоже выдаёт нули
(предпочитает предсказывать не спам).

Эти нули очень часто совпадают.

Чтобы получить более подходящую метрику, вычислим матрицу ошибок.


```python
from collections import Counter

def report_result(mes, pred):
    if mes == 1:
        if pred == 1:
            return 'tp'  # True positive prediction
        else:
            return 'fn'  # False negative
    else:
        if pred == 1:
            return 'fp'  # False positive
        else:
            return 'tn'  # True negative

check = [report_result(m, p) for m, p in zip(messages, predicts)]

# Count results of predictions
cm = Counter(check)

# Print confusion matrix
print(f"tp={cm['tp']:10}", f"  fp={cm['fp']:10}")
print(f"fn={cm['fn']:10}", f"  tn={cm['tn']:10}")
```

Мы видим, что истинно отрицательных предсказаний чрезвычайно много — как уже упоминалось выше, и сообщения,
и предсказания содержат много нулей, поэтому они очень часто совпадают.

Это означает, что значение TN нужно исключить, поскольку оно велико просто случайно, из-за сильной несбалансированности данных.

Вместо accuracy рассматривают precision и recall, а также их комбинацию.

Precision — это доля истинно положительных предсказаний в общем числе положительных предсказаний.

$$
	\text{Precision} = \frac{	\text{TP}}{	\text{TP} + 	\text{FP}}
$$

Recall — это доля истинно положительных предсказаний в общем числе положительных случаев.

$$
	\text{Recall} = \frac{	\text{TP}}{	\text{TP} + 	\text{FN}}
$$

По этим двум значениям вычисляют их гармоническое среднее, которое называется F1-score:

$$
	\text{F1-score} = 2\frac{	\text{Precision} 	\times 	\text{Recall}}{	\text{Precision} + 	\text{Recall}}
$$

Вычислим эти метрики для нашей модели.


```python
precision = cm['tp'] / (cm['tp'] + cm['fp'])
recall = cm['tp'] / (cm['tp'] + cm['fn'])
f1_score = 2 * precision * recall / (precision + recall)

print(f"precision={precision:.3f}")
print(f"recall={recall:.3f}")
print(f"f1_score={f1_score:.3f}")
```

Мы видим, что все три значения очень малы, что соответствует нашей интуиции: наш спам-фильтр не так хорош, как показывает accuracy.


В некоторых случаях precision или recall могут быть важнее друг друга.

Представим, например, что классификатор должен выявлять некоторое заболевание у пациентов.

Положительный результат означает, что у пациента есть болезнь, а отрицательный — что пациент здоров.

В этом случае recall важнее, потому что нам нужно, чтобы классификатор выявил как можно больше действительно больных пациентов.

Другой пример — рекомендательная система. Предсказание классификатора считается положительным, когда рекомендация релевантна, и отрицательным — для нерелевантных рекомендаций.

В этом случае нам нужна высокая precision: большинство положительных рекомендаций должны быть релевантными.

Итак, относительная важность precision и recall должна определяться постановкой задачи.

Ошибочно классифицировать больного человека как здорового и здорового как больного — это разные по цене ошибки.

Именно поэтому F1-score как единственную характеристику классификатора следует использовать осторожно: он приписывает recall и precision одинаковые веса.

Одно из возможных решений — использовать обобщённый F-score (также называемый Fbeta-score), у которого есть дополнительный параметр, изменяющий относительные веса precision и recall:

$$
	\text{F-score} = (1+\beta^2)\frac{	\text{Precision} 	\times 	\text{Recall}}{(\beta^2 \cdot 	\text{Precision}) + 	\text{Recall}}
$$

При $\beta=1$ общий F-score превращается в F1-score.

Меньшее значение beta, например 0.5, придаёт больший вес precision и меньший — recall, тогда как большее значение beta, например 2.0, уменьшает вес precision и увеличивает вес recall при вычислении оценки.


У всех трёх рассмотренных метрик есть свои проблемы. Accuracy плохо работает на несбалансированных данных. Precision и recall асимметричны, то есть игнорируют часть информации. F1-score объединяет их, но неясно, какие веса следует назначать каждой из них.

В некоторых случаях, как уже отмечалось выше, достаточно precision или recall.

Но что делать, если важны оба класса и истинные предсказания для обоих классов очень существенны? Есть ещё один способ объединить результаты бинарной классификации.

Он основан на идее, что истинный класс и предсказанный класс можно рассматривать как две бинарные переменные.

Мера качества предсказания — это коэффициент корреляции между ними.

Ранее мы рассматривали коэффициент корреляции Пирсона. Но тогда мы анализировали временные ряды. Теперь у нас есть две бинарные случайные величины, а TP, TN, FP, FN — это их вероятности.

Чем выше корреляция между истинными и предсказанными значениями, тем лучше предсказание.

Это называется phi-коэффициентом, или коэффициентом корреляции Мэттьюса.


$$
MCC=\frac{TP \times TN-FP \times FN}{\sqrt{(TP+FP)(TP+FN)(TN+FP)(TN+FN)}}
$$


Когда классификатор идеален (FP = FN = 0), значение MCC равно 1, что указывает на идеальную положительную корреляцию.

Напротив, когда классификатор всегда ошибается (TP = TN = 0), мы получаем значение -1, что соответствует идеальной отрицательной корреляции.

В этом случае можно просто обратить результат классификатора и получить идеальный классификатор.

На самом деле значение MCC всегда находится между -1 и 1, а 0 означает, что классификатор не лучше случайного подбрасывания честной монеты.

MCC также полностью симметричен, поэтому ни один класс не важнее другого; если поменять местами положительный и отрицательный классы, мы всё равно получим то же самое значение.

MCC учитывает все четыре значения в матрице ошибок, и высокое значение (близкое к 1) означает, что оба класса предсказываются хорошо, даже если один класс представлен непропорционально мало (или, наоборот, слишком сильно).

Хотя формула для MCC довольно проста, её прямое вычисление, вероятно, завершится неудачей.

Обратите внимание на большое число умножений в знаменателе и числителе. Получающиеся значения будут очень большими!

Но после деления мы получим разумное значение.

Поэтому для вычисления этого значения сначала нужно правильно масштабировать числитель и знаменатель. По этой причине мы воспользуемся готовой функцией из модуля sklearn.


```python
from sklearn.metrics import matthews_corrcoef
MCC = matthews_corrcoef(messages, predicts)
print(f"MCC={MCC:10}")
```

Обратите внимание, что MCC оценивает наш глупый спам-классификатор как очень-очень плохой.

Как уже упоминалось выше, значение, очень близкое к нулю, показывает, что классификатор просто подбрасывает монету.


## Пример обучения модели, недообучения и переобучения


Прежде чем рассматривать пример, обсудим некоторые математические понятия.

Многочлен — это выражение следующего вида:

$$
y = a_0 + a_1 x^1 + a_2 x^2 + a_3 x^3 + \ldots
$$

Наибольшая степень называется степенью многочлена:

$$
y = 2 \text{ :  degree 0}
$$
$$
y = 3-4x \text{ :  degree 1}
$$
$$
y = 6+2x+3x^2 \text{ :  degree 2}
$$

Многочлен можно рассматривать как функцию. В него можно подставить значение $x$ и получить соответствующее значение $y$.

Например, рассмотрим многочлен второй степени.
$$
y = 2 + x^2
$$

Если $x=2$, то соответствующее $y$ равно 6, если $x=4$, то $y=18$ и так далее.

Многочлены можно изображать в виде графиков функций.

В `numpy` есть подмодуль `polynomial`, который предоставляет класс `Polynomial` для работы с многочленами.

Рассмотрим многочлен четвёртой степени

$$
y = 4 + 2 x - 4 x^2 -2 x^3 + 5 x^4
$$


```python
import numpy as np
from numpy.polynomial import Polynomial as P

# We create a polynomial by passing to the class constructior P its coefficients
poly1 = P([4, 2, -4, -2, 5])

# Print the created polynomial
print("y =", poly1)
print("degree =", poly1.degree())

# Check how it works: create list of x values and conpute the corresponding y
x = np.array([-3.8, 4.4, 0.55, 1])
y = poly1(x)
print("x =", x)
print("y =", y)
```

Теперь построим его график.


```python
import matplotlib.pyplot as plt

# This functions runs x in the most appropriate rannge and compute the corresponding y
xp, yp = poly1.linspace()

fig, ax = plt.subplots()
ax.plot(xp, yp)
ax.set_xlabel("x")
ax.set_ylabel("y");
```

Многочлены можно использовать для аппроксимации данных.

Аппроксимация означает поиск такого многочлена, для которого сумма квадратов расстояний от точек данных до кривой многочлена минимальна.

![poly_fit.svg](https://raw.githubusercontent.com/kupav/data-sc-intro/refs/heads/main/fig/poly_fit.png)

Подгонка данных выполняется методом `.fit`.

Ниже мы будем аппроксимировать набор данных многочленами различных степеней.

Эти многочлены будут нашими моделями.

Параметры моделей — коэффициенты многочленов.

Обучение выполняется тогда, когда мы подгоняем многочлен по данным.

Поиск наилучшей аппроксимации данных с помощью некоторого класса функций, например многочленов, называется регрессионным анализом.


Представим, что некоторый природный процесс описывается многочленом.
$$
y = -1 + 1.5 x + 0.1 x^2 - 1.5 x^3
$$


```python
# fun_orig is the true dependence from nature that we will try to recover
from numpy.polynomial import Polynomial as P
import matplotlib.pyplot as plt

fun_orig = P([-1, 1.5, 0.1, -1.5])
print("fun_orig")
print("y =", fun_orig)

xp, yp = fun_orig.linspace()
fig, ax = plt.subplots()
ax.plot(xp, yp)
ax.set_xlabel("x")
ax.set_ylabel("y");
```

Будем считать, что эта формула неизвестна.

Пытаясь её восстановить, мы можем сделать некоторые измерения: значения $x$ и соответствующие им значения $y$.

К сожалению, измерения не являются точными, они искажены шумом.


```python
import matplotlib.pyplot as plt
import numpy as np
rng = np.random.default_rng()

n_total = 30
data = np.zeros((n_total, 2))

# Measured values of x. Simulate them by uniform random numbers
xs = rng.uniform(-1, 1, size=(n_total,))

# Measured values of are computed from xs and spoiled by noise. 
ys = fun_orig(xs) + 0.1 * rng.normal(size=(n_total,))

fig, ax = plt.subplots()
ax.scatter(xs, ys)
ax.set_xlabel("x")
ax.set_xlabel("y");
```

Соберём все данные в набор данных. Значения $x$ составляют первый столбец, значения $y$ — второй столбец.


```python
data = np.array([xs, ys]).T
print(data)
```

Теперь создадим модели машинного обучения, которые будут пытаться восстановить `fun_orig`

Сначала нам нужно разбить набор данных на обучающую, валидационную и тестовую части.

Вот функция для этого.


```python
def split_dataset(data, p_train, p_valid, shuffle=True):
    """Split dataset into train, validation and test parts.
    p_train and p_valid must be within [0, 1]
    """
    assert 0 < p_train + p_valid < 1 
    n_tot = len(data)
    n_train = round(p_train * n_tot)
    n_valid = round(p_valid * n_tot)
    n_test = n_tot - n_train - n_valid
    
    if shuffle:
        # Shuffle data if needed
        rng.shuffle(data, axis=0)
        
    # Extract train dataset
    data_train = data[:n_train]
    # Validation dataset
    data_valid = data[n_train:n_train+n_valid]
    # Test dataset
    data_test = data[n_train+n_valid:n_train+n_valid+n_test]
    return data_train, data_valid, data_test
```

Выполним разбиение.

Возьмём 40\% для обучения и по 30\% для валидации и тестирования.


```python
p_train, p_valid = 0.4, 0.3

data_train, data_valid, data_test = split_dataset(data, p_train, p_valid)

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(12, 3))
axs[0].scatter(data_train[:,0], data_train[:,1], color='C0')
axs[0].set_title(label=f"training, size={len(data_train)}")

axs[1].scatter(data_valid[:,0], data_valid[:,1], color='C1')
axs[1].set_title(f"validation, size={len(data_valid)}")

axs[2].scatter(data_test[:,0], data_test[:,1], color='C2')
axs[2].set_title(f"test, size={len(data_test)}")

for ax in axs:
    ax.grid();
```

Теперь начнём обучать модели.

Моделью является многочлен, и разные модели имеют разные степени многочлена.

Обучение очень простое. Всё делается за один шаг: мы выполняем подгонку многочлена.

Заметьте, что рассмотрение многочленов степени выше 1 означает использование приёма со степенями:

Наш исходный набор данных имеет только два признака: значения $x$ и значения $y$.

И мы добавляем больше признаков, учитывая более высокие степени $x$.


```python
from numpy.polynomial import Polynomial as P

# Max degree of the polynomials
max_degree = 12

models = []
for deg in range(max_degree):
    model = P.fit(data_train[:, 0], data_train[:, 1], deg)
    print("deg=", deg, ":", model)
    models.append(model)
```

Чтобы оценить качество этих моделей, нам нужно подать значения $x$ в каждую из них и вычислить соответствующие значения $y$.

Затем мы сравним вычисленные $y$ с теми, которые содержатся в наборе данных.

Для оценки различия будем использовать метрику MSE.


```python
def mse(a, b):
    """Mean squared error"""
    return np.mean((a-b)**2)
```

Функция, вычисляющая ошибку аппроксимации, выглядит так:


```python
def approx_error(model, data):
    """Approximation error"""
    xs = data[:, 0]
    ys_true = data[:, 1]
    ys_pred = model(xs)
    return mse(ys_true, ys_pred)
```

Посмотрим, как многочлены аппроксимируют наши данные.

Построим графики и вычислим ошибки аппроксимации, которые будут служить оценками качества.


```python
fig, axs = plt.subplots(nrows=max_degree//3, ncols=3, figsize=(12, (max_degree//3)*2.5))

data_plot = data_train
for model, ax in zip(models, axs.reshape(-1)):
    xp_orig, yp_orig = fun_orig.linspace()
    xp, yp = model.linspace()
    ax.scatter(data_plot[:, 0], data_plot[:, 1], color='C0');
    ax.plot(xp_orig, yp_orig, color='C0')  # original dependence that we try to recover
    ax.plot(xp, yp, color='C1')
    error = approx_error(model, data_plot)
    ax.set_title(f"deg={model.degree()}, error={error:.3}")
    ax.grid()
    
plt.tight_layout()
```

Мы видим, что многочлен 11-й степени аппроксимирует обучающие данные идеально.

Ошибка фактически равна нулю. Небольшое ненулевое значение появляется только из-за численных ошибок.

Но что происходит с валидационными данными?

Это данные, которых наши модели никогда не видели.


```python
fig, axs = plt.subplots(nrows=max_degree//3, ncols=3, figsize=(12, (max_degree//3)*2.5))

data_plot = data_valid
for model, ax in zip(models, axs.reshape(-1)):
    xp_orig, yp_orig = fun_orig.linspace()
    xp, yp = model.linspace()
    ax.scatter(data_plot[:, 0], data_plot[:, 1], color='C0');
    ax.plot(xp_orig, yp_orig, color='C0')  # original dependence that we try to recover
    ax.plot(xp, yp, color='C1')
    error = approx_error(model, data_plot)
    ax.set_title(f"deg={model.degree()}, error={error:.3}")
    ax.grid()
    
plt.tight_layout()
```

Теперь всё иначе.

Ошибка аппроксимации для многочлена 11-й степени огромна.

Для лучшей визуализации соберём ошибки аппроксимации в массивы и построим их графики.


```python
import matplotlib.pyplot as plt

degrees = [model.degree() for model in models]
error_train = [approx_error(model, data_train) for model in models]
error_valid = [approx_error(model, data_valid) for model in models]

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))

# Apprximation error vs degree of the model
ax = axs[0]
ax.plot(degrees, error_train, '-*', label='train')
ax.plot(degrees, error_valid, '-*', label='valid')
ax.set_title("Log scale")
ax.legend()
ax.set_yscale('log')
ax.set_ylim([1e-4, 1])
ax.grid()

# Same but y scale is linear 
ax = axs[1]
ax.plot(degrees, error_train, '-*', label='train')
ax.plot(degrees, error_valid, '-*', label='valid')
ax.set_title("Linear scale")
ax.legend()
ax.legend()
ax.set_ylim([0, 0.1])
ax.grid()
```

Мы видим, что модели низких степеней 0 и 1 недообучены.

Они плохо работают и на обучающих, и на валидационных данных.

Модели высоких степеней переобучены. У модели 11-й степени ошибка на обучающих данных равна 0, потому что она просто запомнила их.

Если показать ей ранее не встречавшиеся данные, это приводит к огромной ошибке.

Оптимальные модели имеют степень около 3 — ту же, что и исходный многочлен.

Мы использовали обучающие данные для вычисления параметров модели. Это коэффициенты многочлена.

Эти данные нельзя использовать для вычисления итоговой оценки модели.

Валидационные данные использовались для выбора лучшей модели. Это тоже своего рода обучение: мы берём модель, которая лучше всего работает на валидационных данных.

Валидационные данные тоже нельзя использовать для вычисления итоговой оценки.

Именно поэтому нам нужны тестовые данные.

Они показываются выбранной модели только один раз, и после этого никакого выбора модели больше не выполняется.

Выберем модель степени 3 как результат нашего обучения и найдём её итоговую оценку на тестовых данных.


```python
import matplotlib.pyplot as plt

the_model = models[3]
print("the_model:\n", the_model)

fig, ax = plt.subplots()

xp_orig, yp_orig = fun_orig.linspace()
xp, yp = the_model.linspace()
ax.scatter(data_test[:, 0], data_test[:, 1], color='C0');
ax.plot(xp_orig, yp_orig, color='C0')  # original dependence that we try to recover
ax.plot(xp, yp, color='C1')

error = approx_error(the_model, data_test)
ax.set_title(f"deg={the_model.degree()}, error={error:.3}")
ax.grid()
```

Обсудим, почему переобучение возникает у многочленов высокой степени.

У этих многочленов слишком много параметров по сравнению с размером набора данных.

Модель может легко запомнить все входы, не находя их обобщения.

Поэтому с переобучением можно бороться либо уменьшая число параметров, то есть рассматривая многочлены меньшей степени.

Либо можно (если это возможно) найти больше данных. Тогда модели не смогут запомнить все эти данные и будут вынуждены искать обобщение.

Ещё один способ избежать переобучения называется регуляризацией.

Применение регуляризации к параметрам модели означает ограничение диапазонов их изменения.

Идея та же самая: сжимая диапазон допустимых изменений, мы запрещаем простое запоминание данных.

Посмотрим ещё раз на наши модели:


```python
for model in models:
    print("deg=", model.degree(), '\n', model, '\n')
```

Модели высокой степени, то есть сильно переобученные, действительно имеют очень большие параметры.

Регуляризация означает, что при поиске оптимальных параметров мы добавляем дополнительный штраф за величины параметров, требуя, чтобы они были не слишком большими.

Этот метод называется гребневой регрессией.
