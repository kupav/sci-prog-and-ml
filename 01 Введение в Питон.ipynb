# Введение в Питон

## Задание 1: Вывод и базовые переменные
Напишите программу, которая создает три переменные: `фамилия` (строка), `год_рождения` (целое число) и `средний_балл` (число с плавающей точкой). Затем выведите на экран строку в формате: "Студент [Фамилия], родившийся в [Год], имеет средний балл [Балл]".

## Задание 2: Операторы присваивания с модификацией
Создайте переменную `counter` со значением 10. Используя оператор `-=`, уменьшите ее значение на 3. Используя оператор `*=`, умножьте ее значение на 2. Выведите финальное значение.

## Задание 3: Исследование типов переменных
Создайте переменные: `a = 42`, `b = 3.14159`, `c = "Математика"`, `d = 2+3j`. Используя функцию `type()`, выведите тип каждой переменной на экран.

## Задание 4: Приведение типов (int, float, str)
Программа просит пользователя ввести два числа (как строки). Преобразуйте их к типу `float`, вычислите их сумму и произведение. Выведите результаты, указав тип каждого результата.

## Задание 5: Удаление переменной и проверка
Создайте переменную `temp_data = 100`. Выведите ее на экран. Затем удалите эту переменную с помощью оператора `del`. Попробуйте вывести ее снова, обработав возможную ошибку с помощью `try...except NameError`.

## Задание 6: Приоритет операций
Дано выражение: `10 + 2 * 3 ** 2`. Вычислите его "в уме", записав ответ в переменную `my_answer`. Затем попросите Python вычислить это выражение и сохранить в `python_answer`. Сравните их и выведите оба результата.

## Задание 7: Изменение приоритета скобками
Возьмите выражение из задания 6. Используя **только круглые скобки**, измените его так, чтобы порядок вычислений стал: сложение, умножение, возведение в степень. Вычислите и выведите новый результат.

## Задание 8: Комплексные числа
Создайте два комплексных числа: `z1 = 1 + 2j` и `z2 = 3 - 4j`. Вычислите и выведите их сумму, разность и произведение. Выведите действительную и мнимую часть произведения.

## Задание 9: Деление с остатком и целочисленное деление
Для чисел `a = 17` и `b = 5` вычислите и выведите результат обычного деления (`/`), целочисленного деления (`//`) и взятия остатка (`%`). Проверьте выполнение свойства: `a = b * (a // b) + (a % b)`.

## Задание 10: Логический (булев) тип
Создайте переменные: `x = 10`, `y = 20`, `z = 10`. Вычислите и выведите результаты выражений: `x > y`, `x == z`, `(x < y) and (y > z)`. Объясните в комментарии, какой тип имеют результаты этих выражений.

## Задание 11: Работа с битами (логическое И - &)
Числа `a = 12` (бинарное `1100`) и `b = 10` (бинарное `1010`). Вычислите и выведите результат побитового И (`a & b`). Переведите результат в двоичную систему с помощью `bin()` и выведите.

## Задание 12: Работа с битами (логическое ИЛИ - | и XOR - ^)
Для тех же чисел `a = 12` и `b = 10` вычислите и выведите результаты побитового ИЛИ (`a | b`) и исключающего ИЛИ (`a ^ b`). Выведите результаты в двоичном виде.

## Задание 13: Битовый сдвиг влево (<<)
Число `n = 5` (бинарное `101`). Выполните операцию побитового сдвига влево на 2 позиции (`n << 2`). Объясните в комментарии, как эта операция связана с умножением на степень двойки. Выведите результат в десятичном и двоичном виде.

## Задание 14: Битовый сдвиг вправо (>>)
Число `n = 20` (бинарное `10100`). Выполните операцию побитового сдвига вправо на 2 позиции (`n >> 2`). Объясните в комментарии, как эта операция связана с целочисленным делением на степень двойки. Выведите результат.

## Задание 15: Проверка на четность с помощью битовых операций
Напишите программу, которая запрашивает целое число и с помощью побитовой операции **`&`** проверяет, является ли оно четным.

## Задание 16: Инвертирование битов (~)
Создайте переменную `x = 5` (бинарное `...00000101`). Примените к ней оператор побитового НЕ (`~x`). Выведите результат. Объясните в комментарии, почему результат отрицательный.

## Задание 17: Составное присваивание с битовыми операциями
Начальное значение `mask = 0b0000`. С помощью операторов `|=` и `<<=` установите (т.е. сделайте равными 1) -й и 3-й биты (считая с младшего, с 0). Выведите маску в двоичном и десятичном виде.

## Задание 18: Вычисление выражения со смешанными операторами
Вычислите значение выражения: `(15 % 4) * (2 ** 3 + 1) - (9 // 2)`. Запишите его в переменную и выведите. Разберите по шагам порядок операций в комментарии.

## Задание 19: Обмен значений переменных (без временной переменной)
Обменяйте значения двух целых переменных `p` и `q` **без использования третьей переменной**. Выведите значения до и после. Напишите два варианта - используя арифметические операции и используя битовые операции.

## Задание 20: Сумма цифр двузначного числа (с использованием деления и остатка)
Напишите программу, которая для заданного двузначного целого числа `num` (например, 57) вычисляет и выводит сумму его цифр. Используйте только операции `//` и `%`.
