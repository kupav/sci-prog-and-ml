## Задание 12: Накопление погрешности в итеративных процессах
Интеграл
$$I_n = \int_0^1 \frac{x^n}{x+5} dx$$
может быть вычислен по рекуррентной формуле
$$I_n = \frac{1}{n} - 5 I_{n-1}, \quad I_0 = \ln\frac{6}{5}$$
При этом последовательность $I_n$ убывающая (как это доказать?) поэтому для больших $n$ можно приближённо считать $I_n=0$.

Рекуррентная формула численно неустойчива, а вычисления по обратной формуле - устойчивы. Чтобы в этом убедиться, выполните следующие вычисления.
1. Вычислите интеграл непосредственно, используя численную процедуру `scipy.integrate.quad`. Эта процедура находит интеграл численно с контролем погрешности и с достаточно высокой точностью. Мы будем считать найденные ею значения точными.
2. Вычислите $I_n$ для $n = 0,1,\dots,25$ с помощью рекуррентного соотношения, начиная с точного $I_0$.
3. Вычислите те же значения "с конца" (обратная рекурсия), используя грубую оценку для большого $n$, положив $I_{30}=0$ и двигаясь назад.
4. Напечатайте таблицу точных значений $I_n$, значений вычисленных при прямом и обратном проходах, а также соответствующие абсолютные погрешности.
5. Сравните результаты и убедитесь, что прямые итерации очень быстро становятся неточными.
