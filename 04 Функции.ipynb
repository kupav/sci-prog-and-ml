# Задания по созданию и использованию функций Python

## Задание 1: Объявление простой функции
Создайте функцию `circle_area(radius)`, которая принимает радиус круга и возвращает его площадь (π × r²). Используйте значение π = 3.14159.

## Задание 2: Функция с параметрами по умолчанию
Создайте функцию `power(base, exponent=2)`, которая возвращает `base` в степени `exponent`. Параметр `exponent` должен быть по умолчанию равен 2.

## Задание 3: Именованные аргументы
Создайте функцию `rectangle_properties(length, width)`, которая возвращает кортеж (периметр, площадь) прямоугольника. Продемонстрируйте вызов функции с использованием именованных аргументов в разном порядке.

## Задание 4: Функция с переменным числом позиционных аргументов (`*args`)
Создайте функцию `average(*numbers)`, которая принимает произвольное количество чисел и возвращает их среднее арифметическое.

## Задание 5: Функция с переменным числом именованных аргументов (`**kwargs`)
Создайте функцию `print_student_info(**info)`, которая принимает произвольное количество именованных аргументов (характеристик студента) и выводит их в формате "ключ: значение".

## Задание 6: Лямбда-функция для проверки чётности
Создайте лямбда-функцию `is_even`, которая принимает число и возвращает `True`, если число чётное, и `False` в противном случае.

## Задание 7: Лямбда-функция в сортировке
Дан список кортежей `points = [(1, 3), (4, 1), (2, 2), (0, 0)]`. Отсортируйте его по расстоянию от точки до начала координат (0, 0), используя лямбда-функцию в качестве ключа сортировки.

## Задание 8: Функция, возвращающая лямбда-функцию
Создайте функцию `multiplier(n)`, которая возвращает лямбда-функцию, умножающую свой аргумент на `n`.

## Задание 9: Простой генератор арифметической прогрессии
Создайте функцию-генератор `arithmetic_sequence(start, step)`, которая бесконечно генерирует члены арифметической прогрессии, начиная с `start` и с шагом `step`.

## Задание 10: Генератор чисел Фибоначчи до предела
Создайте функцию-генератор `fibonacci_limit(limit)`, которая генерирует числа Фибоначчи, не превышающие заданный `limit`.

## Задание 11: Декоратор для замера времени выполнения
Создайте декоратор `timer`, который измеряет время выполнения декорируемой функции и выводит его в секундах.

## Задание 12: Декоратор для кэширования (мемоизации)
Создайте декоратор `cache`, который сохраняет результаты вызовов функции для определённых аргументов и возвращает сохранённое значение при повторном вызове с теми же аргументами.

## Задание 13: Рекурсивное вычисление факториала
Создайте рекурсивную функцию `factorial_recursive(n)` для вычисления факториала числа n.

## Задание 14: Рекурсивное нахождение НОД (алгоритм Евклида)
Создайте рекурсивную функцию `gcd_recursive(a, b)` для нахождения наибольшего общего делителя двух чисел, используя алгоритм Евклида.

## Задание 15: Рекурсивный обход вложенных структур
Создайте рекурсивную функцию `deep_sum(nested_list)`, которая вычисляет сумму всех чисел во вложенных списках произвольной глубины.

## Задание 16: Глобальная переменная в функции
Создайте функцию `counter()` без аргументов, которая при каждом вызове увеличивает глобальную переменную `count` на 1 и возвращает её текущее значение.

## Задание 17: Замыкание с изменяемым состоянием
Создайте функцию `make_counter()`, которая возвращает функцию-счётчик. Каждый вызов возвращённой функции должен увеличивать внутренний счётчик и возвращать его значение. Не используйте глобальные переменные.

## Задание 18: Функция с комбинацией `*args` и `**kwargs`
Создайте функцию `print_args(*args, **kwargs)`, которая:
1. Выводит все позиционные аргументы с префиксом "Позиционные:"
2. Выводит все именованные аргументы с префиксом "Именованные:"
3. Возвращает кортеж (количество позиционных аргументов, количество именованных аргументов)

## Задание 19: Генератор с условием остановки и параметрами
Создайте функцию-генератор `primes_generator(limit)`, которая генерирует простые числа до заданного предела, используя решето Эратосфена или простую проверку делителей.

## Задание 20: Декоратор с параметрами
Создайте декоратор `repeat(n)`, который вызывает декорируемую функцию `n` раз и возвращает список результатов всех вызовов.
