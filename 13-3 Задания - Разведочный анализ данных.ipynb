## Учебный датасет

Ниже создаётся искусственный датасет о клиентах сервиса доставки еды.  
Целевая переменная: **made_order_again** — сделал ли клиент повторный заказ.

В таблицу специально добавлены:
- пропуски;
- некорректные значения;
- разные записи одной и той же категории;
- выброс.

Это позволяет потренировать базовые шаги EDA без сложных операций над данными.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display

np.random.seed(123)

n = 150

age = np.random.randint(18, 66, size=n).astype(float)
monthly_orders = np.clip(np.random.normal(6, 2.2, size=n), 0, 18)
avg_order_value = np.clip(np.random.normal(780, 180, size=n), 150, 2000)
delivery_time = np.clip(np.random.normal(36, 9, size=n), 10, 90)
support_calls = np.random.poisson(1.8, size=n)
city_zone = np.random.choice(["Center", "North", "South"], size=n, p=[0.4, 0.3, 0.3]).astype(object)
app_type = np.random.choice(["Android", "iOS", "Web"], size=n, p=[0.45, 0.35, 0.20]).astype(object)

score = (
    0.22 * monthly_orders
    + 0.0022 * avg_order_value
    - 0.035 * delivery_time
    - 0.18 * support_calls
    + (city_zone == "Center") * 0.25
    + (app_type == "iOS") * 0.15
    + np.random.normal(0, 0.45, size=n)
)
made_order_again = (score > 1.4).astype(int)

df_source = pd.DataFrame({
    "age": age.round(0),
    "monthly_orders": monthly_orders.round(1),
    "avg_order_value": avg_order_value.round(0),
    "delivery_time": delivery_time.round(1),
    "support_calls": support_calls,
    "city_zone": city_zone,
    "app_type": app_type,
    "made_order_again": made_order_again
})

# Добавляем искажения
df_source.loc[[7, 92], "age"] = -5
df_source.loc[[11, 78], "avg_order_value"] = np.nan
df_source.loc[34, "avg_order_value"] = -300
df_source.loc[55, "delivery_time"] = 210
df_source.loc[[21, 99, 132], "app_type"] = np.nan
df_source.loc[[14, 83], "city_zone"] = "center"
df_source.loc[[47, 126], "city_zone"] = "South "
df_source.loc[108, "support_calls"] = 32

df_source.to_csv("eda_practice_dataset_3.csv", index=False)

print("Файл eda_practice_dataset_3.csv создан.")
display(df_source.head())
````

# Задание 10. Сравнение модели на «сыром» и очищенном датасете

## Что нужно сделать
1. Попробуйте обучить простую модель на исходном датасете `df`.
2. Затем обучите такую же модель на очищенном датасете `df_clean`.
3. Сравните результаты и сделайте вывод, зачем нужен EDA.

## Зачем это нужно
Это итоговое задание, где видно, что EDA влияет не только на аккуратность таблицы, но и на возможность построить модель.

Логика такая:
- сначала пробуем обучить модель на данных «как есть»;
- затем используем очищенные данные;
- сравниваем, изменилось ли обучение и качество.

## Почему это важно
Такой эксперимент показывает, что EDA:
- делает обучение модели возможным;
- уменьшает число технических ошибок;
- помогает получить более понятные признаки.

## Подсказки по Pandas и машинному обучению
- признаки `X` нужно отделить от целевой переменной `y`;
- категориальные признаки можно перевести в числа через `pd.get_dummies(...)`;
- данные обычно делят на обучающую и тестовую части;
- после этого создают модель, обучают её и оценивают качество.

Если вы ещё не изучали модели, используйте такой шаблон:

1. Подготовка признаков и целевой переменной:
```python
X = df_clean.drop(columns=["made_order_again"])
y = df_clean["made_order_again"]
```

2. Преобразование категориальных столбцов:
```python
X = pd.get_dummies(X, drop_first=False)
```

3. Разбиение на обучающую и тестовую части:
```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

4. Создание и обучение модели:
```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```
Увеличьте `max_iter` если модель не будет успевать сойтись.

5. Получение предсказаний и оценка качества:
```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print(accuracy_score(y_test, y_pred))
```
